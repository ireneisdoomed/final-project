```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Train

```python
train = pd.read_csv("../data/sign_mnist/sign_mnist_train.csv")
```

```python
print(train.shape)

# 27455 de 28x28 pixeles
```

```python
train.head()
```

```python
#Columna Ground Truth a letras

newLabels = {0: "A", 1: "B", 2: "C", 3: "D", 4: "E", 5: "F", 6: "G",
            7: "H", 8: "I", 10: "K", 11: "L", 12: "M", 13: "N",
            14: "O", 15: "P", 16: "Q", 17: "R", 18:"S",
            19: "T", 20:"U", 21:"V", 22:"W", 23:"X",
            24:"Y"}

train["label"] = train["label"].map(newLabels)
```

```python
X_train = train.drop("label", axis=1)
y_train = train["label"]
```

### Test

```python
test = pd.read_csv("../data/sign_mnist/sign_mnist_test.csv")
```

```python
print(test.shape)
```

```python
test["label"] = test["label"].map(newLabels)
```

```python
X_test = test.drop("label", axis=1)
y_test = test["label"]
```

```python
# 7172 de 28x28 pixeles

print(X_test.shape)

# X: feature vectors
# y: GT
```

## Â¿Con quÃ© gestos me quedo?

### Entreno con RF y hago matriz de confusiÃ³n

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf = RandomForestClassifier(max_depth=20, random_state=0).fit(X_train, y_train)

print(accuracy_score(y_test, clf.predict(X_test)))
```

```python
from sklearn.metrics import confusion_matrix

y_pred = clf.predict(X_test)
plt.figure(figsize =(30,17))
cm = confusion_matrix(y_test, y_pred)


sns.heatmap(confusion_matrix(y_test, y_pred), annot=True).set(xlabel='GROUND TRUTH', ylabel='PREDICTION', xticklabels=set(y_test), yticklabels=set(y_pred));

```
